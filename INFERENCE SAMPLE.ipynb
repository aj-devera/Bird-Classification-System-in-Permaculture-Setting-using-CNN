# Prediction Function

```python
def predict_species(image_file):
    img = load_img(image_file, target_size=(224,224))
    x = img_to_array(img)
    x = np.expand_dims(x, axis=0)
    x = x/255.0
    values = CNN_model.predict(x)
    idx = np.argmax(values[0])
    predict = species[idx]
    
    return predict
```

# Images of the Mel Spectrogram

```python
# Filepaths of the images
true_image = os.path.join(images_filepath, "true image.png")
test_image = os.path.join(images_filepath, "test image.png")

# Showing these images
fig, axs = plt.subplots(1,2, figsize=(15,10))
img = Image.open(true_image)
test_img = Image.open(test_image)
axs[0].imshow(img)
axs[0].set_title('True Image', fontsize=15)
axs[1].imshow(test_img)
axs[1].set_title('Test Image', fontsize=15)
plt.show()
```

Predicting the what bird species from the test image on the right using a CNN model and the mel spectrogram

```python
# Loading the CNN model
CNN_model = keras.models.load_model(CNN_model_filepath + "/custom_model_5 seconds bandpass only.h5")

# Predicting the right species
prediction = predict_species(test_image)
print(f"\nThe Predicted Species is a {prediction}")
```
